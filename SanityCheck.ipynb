```python
# (vals/counts*100) - percent
def describe(arr):
    print('''
    %.2e Mean
    %.2e Median
    %.2e Standard Deviation
    %.2e Max
    %.2e Min
    %.2f Non-Zero Percent 
    '''%(np.mean(arr),np.median(arr),np.std(arr),
         np.max(arr),np.min(arr),100*(np.count_nonzero(arr)/np.multiply(*arr.shape))))
```

```python
describe(counts)
```

```python
describe(vals)
```

```python
describe(percent)
```
